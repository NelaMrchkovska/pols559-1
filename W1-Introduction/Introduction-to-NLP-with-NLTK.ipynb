## Counting Vocabulary 

```python
# len gives us the length of a text from start to finish, in terms of the words and punctuation symbols that appear
# 44764 "tokens"
len(text3)
```

```python
# The vocabulary of a text is just the set of tokens that it uses, since in a set, all duplicates are collapsed together. 
sorted(set(text3))
```

```python
len(set(text3))
# Although it has 44,764 tokens, this book has only 2,789 distinct words, or "word types."
```

```python
len(set(text3)) / len(text3)
# This shows the lexical richness of the text - the number of distinct words is just 6% of the total number of words
```

```python
text3.count("smote")
# We can count how often a word occurs in a text
```

* What if we want to repeat such calculations on several texts? 

```python
def lexical_diversity(text):
    return len(set(text)) / len(text)
```

```python
lexical_diversity(text3) # it works! 
```

## Frequency Distribution

```python
fdist1 = FreqDist(text1)
print(fdist1)
```

```python
fdist1.most_common(50)
# This gives us a list of the 50 most frequently occurring types in the text
```

```python
fdist1.plot(50, cumulative=True)
# These 50 words account for nearly half the book!
```

## Fine-grained Selection of Words

* What if we want to find the words from the vocabulary of the text that are more than 15 characters long?

```python
V = set(text1) # create the vocabulary for text1
long_words = [w for w in V if len(w) > 15] # For each word w in the vocabulary V check if its length is greater than 15
sorted(long_words)
```

* What if the long words cannot typify a text (like yuuuuuuuuuuuuuummmmmmmmmmmm)? 

```python
# We need to find frequently occurring long words
fdist5 = FreqDist(text5)
sorted(w for w in set(text5) if len(w) > 7 and fdist5[w] > 7)
# len(w) > 7 ensures that the words are longer than seven letters, and fdist5[w] > 7 ensures that these words occur more than seven times
```
